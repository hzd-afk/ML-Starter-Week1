```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
```

```python
load_diabetes()
```

```python
df=load_diabetes()
```

```python
dataset=pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns=df.feature_names
```

```python
dataset.head()
```

```python
##independent features and depented features
```

```python
x=dataset
y=df.target
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
##standardizing the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
trainScaler=scaler.fit_transform(X_train)
```

```python
testscaler=scaler.transform(X_test)
```

```python
testscaler
```

```python
#cross validation
from sklearn.model_selection import cross_val_score
```

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
mse
```

```python
np.mean(mse)
```

```python
##prediction
reg_predict=regression.predict(X_test)
```

```python
reg_predict
```

```python
import seaborn as sns
sns.displot(reg_predict-y_test,kind='kde')#we get an assumption that are dataset is good
#high variance bad training
```

```python
from sklearn.metrics import  r2_score
```

```python
score=r2_score(reg_predict,y_test)
```
